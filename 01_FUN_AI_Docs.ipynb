# GitHub アカウントの作成

### ※01_github.pdf にもほぼ同様の内容が書いてある。そちらを見た場合はこの章は飛ばしてもよい。

GitHubとは、プログラムのバージョン管理システムを補助するシステムの1つだ。

複数人で開発するときには必須と言ってもいいから、ここで導入を行う。ProcessingのソースなどもGitHubで管理すると何かと便利なことが多い。

[公式サイト](https://github.com)にアクセスし、sign upする。とりあえずは学内メールで作るとよいだろう。(あとから複数のメールアドレス追加できる)

ついでに sign in もしてしまおう。GitHubから来るメールを確認しverifyすること。

次に[このサイト](https://qiita.com/toshi-click/items/dcf3dd48fdc74c91b409)を参考に、Gitをコマンドラインで使う準備をする。後ほど自分で書くかもしれない。

それらが済んだら、僕のGitHubからサークルで使う資料を[ダウンロード](https://github.com/t4t5u0/2020_fun_ai_docs)する。

![](./fig/01/01_03.png)

リンクをコピーし、`git clone`する。

`git clone` する際の例をあげる。まずはWindows powershell を立ち上げる。

Windowsキー(スタートキー)を押し、アプリを検索できる状態にし、powershell などと打つと良い

`cd`コマンドはchange directry の略で、ディレクトリ(フォルダ)を移動できる。`~/`はホームディレクトリを表す。

```Console
$ cd ~/
$ cd %HOMEPATH% (Windowsの場合)
$ cd Documents
$ git clone https://github.com/t4t5u0/2020_fun_ai_docs.git 
$ cd 2020_fun_ai_docs
$ git branch -a
$ git checkout dev
```

更新したい場合は
```Console
$ cd 2020_fun_ai_docs
$ git pull
```

# Python(Anaconda)の環境構築

注意！すでに環境構築されていて、自分の環境を汚したくない場合は、いい感じにやってください

## Anacondaのインストール
Anacondaとは機械学習やデータサイエンスのためのPythonディストリビューションだ。様々なライブラリがデフォルトでインストールされていたり、複数の環境を用意できたりする。特に、複数の環境を用意できる点が嬉しく、プロジェクトごとに言語やライブラリのバージョンを管理できる。複数のプロジェクトを同時に行うとき、別々の環境が欲しいことが多い。そのために、`conda` というコマンド群を使って、環境を複数用意する。

それでは、環境を作って行く。
まずは[公式サイト](https://www.anaconda.com/distribution/#download-section)からダウンロードする。Python3.7 を選択すること。

![](./fig/01/01_02.png)

途中で一個チェックを入れる。他は特にこだわりが無ければ、そのまま進めて良い。

![](./fig/01/01_09.png)

## VSCode のインストール
[公式サイト]()からWindows 64bit 版をダウンロードする。ダウンロードしたらとりあえず起動してみる。  

VSCodeは拡張機能が豊富で色々なものがある。

- zenkaku
    - 全角スペースを強調してくれる

インストールについては下図参照。検索窓に入れたい拡張機能の名前を入れる

![](./fig/01/01_04.png)

他にも色々な拡張機能があるから、こういった[記事](https://qiita.com/sensuikan1973/items/74cf5383c02dbcd82234)を参考にぜひ自分好みのVSCodeを作って欲しい。

Anaconda powershell を起動して、以下のコマンドを打つ。これで、`circle`という名前のconda環境が作られた。今後、特にこだわりが無ければ、`circle`環境を使うことを推奨する。

## VSCode で jupyter notebook を使う方法

VSCodeでjupyter notebook を使うこともできる。VSCodeの拡張機能でPythonを有効にする

次は、VSCodeで用いるPythonのバージョンを選択する`Ctrl+Shift+p`を押し、
```
Python：select interpreter
>Python 3.7.4 64-bit ('base':conda)
```
を選択  
左下にPython 3.7.4 64-bit ('base':conda) と表示されていれば良い

新しいウィンドウを開く。(`Ctrl+Shift+n`) その後、フォルダを開く(`Ctrl+k Ctrl+o`) から先程`git clone`した `2020_fun_ai_docs`を選択する

VSCode左のエクスプローラーから`01_FUN_AI_Docs.ipynb` を開く

VSCodeのリボンからターミナル(T)→新しいターミナル。もしくは ``Ctrl+Shift+` `` を打つ。(バッククォート(`)は
JISキーボードだとShift+＠)

```Console
$ conda create -n circle
$ conda activate circle
```

右上の`jupyter server: `の欄がNot Started になっていた場合は以下の作業を行う。

## PyTorchのインストール

PyTorchとは、深層学習用のPythonフレームワークだ。FUN AI では、PyTorchを使って演習を進めていく。詳しい使い方などは次回以降に解説していく。
[公式サイト](https://pytorch.org)にアクセスして、欲しいバージョンや環境をクリックするとリンクを生成してくれる。Windows用のものは以下に用意した。他のOSも同じようにダウンロードできる。

先程同様にVSCode内でターミナルを開く。(``Ctrl+Shift+` ``) その後、以下のコマンドを打つ
```Console
$ conda activate cirle
$ conda install pytorch torchvision cudatoolkit=10.1 -c pytorch
```

Google Colabratory(Colab)とは、Googleのサービスの1つで、Web上でnotebook形式のPythonを走らせることができる。

深層学習の学習に必要なGPUを無料で使うことができたり、頼ることもあるだろう。
